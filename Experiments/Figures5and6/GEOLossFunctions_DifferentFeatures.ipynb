```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

# 500

```python
filename='./log/log_500.csv'
df=np.array(pd.read_csv(filename,header=0,sep=';'))

epochs_=df[:200,0]
loss_=df[:200,1]
val_loss_=df[:200,3]

epochs_number=200
loss = loss_
val_loss = val_loss_

epochs = range(epochs_number)

plt.plot(epochs, loss, 'bo', label='Training Loss')
plt.plot(epochs, val_loss, 'r', label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
