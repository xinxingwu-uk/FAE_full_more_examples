```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.style.use('seaborn-dark')

selected_features = [0,2,4,6,8,10]

fig = plt.figure(figsize = (10,6))

ax1 = fig.add_subplot(111)
ax1.set_ylabel('Reconstruction error',fontsize=28)
plt.grid(True,linestyle='--',linewidth = 2.5,zorder=-1,axis="y")
plt.xticks(selected_features, fontsize = 28)
plt.xlabel('Number of selected genes', fontsize = 28)
plt.ylabel('Reconstruction error', fontsize = 28)

error = [0.3102578944745165,0.29902730801459276,0.2882793171569336,0.2766890695148241,0.26924063796860925,0.2665971278965129]

plt.plot(selected_features, error, marker='*', c=sns.xkcd_rgb["denim blue"], label='Accuracy',ms=25,lw=3)
#plt.text(selected_features[3]-1.8, error[3]+0.005, "0.024", c='black',size = 20, alpha = 0.8)

plt.xticks(selected_features,np.array(["500","600","700","800","900","943"]),fontsize=28)
plt.yticks(fontsize=28)

plt.yticks(np.arange(0.24,0.32,0.025),fontsize=28)
plt.ylim([0.24, 0.32])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.style.use('seaborn-dark')

selected_features = [0,2,4,6,8,10]

fig = plt.figure(figsize = (10,6))

ax1 = fig.add_subplot(111)
ax1.set_ylabel('Reconstruction error',fontsize=28)
plt.grid(True,linestyle='--',linewidth = 2.5,zorder=-1,axis="y")
plt.xticks(selected_features, fontsize = 28)
plt.xlabel('Number of selected genes', fontsize = 28)
plt.ylabel('Linear reconstruction error', fontsize = 28)

error = [0.2997106337341877,0.285859601322965,0.2743231780710275,0.2650351564588178,0.25645862966425415,0.252957551725207]

plt.plot(selected_features, error, marker='*', c=sns.xkcd_rgb["pale red"], label='FAE-selected genes',ms=25,lw=3)
#plt.text(selected_features[3]-1.8, error[3]+0.005, "0.024", c='black',size = 20, alpha = 0.8)
plt.axhline(y=0.26325423264377185, color='g', linewidth = 5,linestyle='dashed',label='943 landmark genes')

#plt.hlines(y=0.26325423264377185, xmin=0, xmax=10, color='g', linewidth = 3.5,linestyle='dashed',label='943 landmark genes')

plt.xticks(selected_features,np.array(["500","600","700","800","900","943"]),fontsize=28)
plt.yticks(fontsize=28)

plt.yticks(np.arange(0.24,0.32,0.02),fontsize=28)
plt.ylim([0.24, 0.31])

plt.legend(fontsize=28)
plt.show()
```
